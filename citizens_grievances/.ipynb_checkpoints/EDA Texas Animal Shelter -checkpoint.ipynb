## Problem Statement

This Notebook focuses on data understanding and preparation for a particular problem and dataset.
The problem and data come from an animal shelter concerned with animal welfare and reducing the risk of animal death. The shelter wants to use the data collected about their animals to build a data analytics solution for death risk prediction to help them in their planning towards improving the welfare of the animals they shelter. The shelter collects some data for each animal they take in (columns in the dataset with keyword "intake") and also records the status of each animal when it left the shelter (columns in the dataset with keyword "outcome").
The target variable to predict is a binary variable called **binary_outcome**. For this variable, the value “1” indicates that after intake, the animal outcome was negative, e.g., "death", while the value "0" indicates that the animal outcome was a positive one, e.g., was adopted or returned to the owner.
The dataset we work with is a sample of the data released by this shelter:
https://data.austintexas.gov/Health-and-Community-Services/Austin-Animal-Center-Outcomes/9t4d-g238

#### 1.1.3.4 Convert to decided data type 

```python
#CATAGORICAL DATA

# Select columns containing categorical data
categorical_columns = df[['Animal ID','Found Location','Name_Intake','Intake Type','Intake Condition','Intake Condition'\
                         ,'Animal Type_Intake','Sex upon Intake','Breed_Intake','Color_Intake'\
                         ,'Name_Outcome','Animal Type_Outcome','Sex upon Outcome','Breed_Outcome'\
                         ,'Color_Outcome','binary_outcome']].columns

# Explicitely convert Target variable binary_outcome from float 64 to int
df['binary_outcome'] = pd.to_numeric(df['binary_outcome'], downcast='integer')

#Invert Terminology for Positive and negative outcome 1:Good 0: Bad
for i in df.index:
    df['binary_outcome'][i] = (df['binary_outcome'][i] + 1) % 2

# Assign data type category to columns listed in categorical_columns
# Convert data type of target variable 'binary_outcome' from float 64 to category
for column in categorical_columns:
    df[column] = df[column].astype('category') 
```

```python
#CONTINUOUS DATA

# Select columns containing continuous data
continous_columns = df[['Age upon Intake', 'Age upon Outcome']].columns

# Represent age in terms of "Number of months"
# Assign data type int64 to columns enlisted in continous_columns
# Glance at data in spreadsheet states that ages of animals are mostly expressed in months; then in years and rarely in weeks and days.
# Unit "weeks" is used as approximation scale 
# Hence, if condition for month is kept first followed by years and weeks; in order to exploit shortcircuiting of if-elif-else loops
# Additionaly, if any cell has value 0 Or value is less than 7 days, it is marked 1 [considering that with present units, animal taken in is of minimum 1 week]
# For invalid values marked UNKNOWN or anything else, random integer 0 is assigned.

for column in continous_columns:
    if pd.api.types.is_string_dtype(df[column].dtypes) :
        for cell in range(df.shape[0]):
            if  pd.isnull(df.iloc[cell][column]):
                df.at[cell,column] = "1 month"
                
            age = str(df.iloc[cell][column]).split(" ")
            
            if (int(age[0])== 0) and ('week' in age[1] or 'month' in age[1] or 'year' in age[1] or 'day' in age[1]):
                df.at[cell,column] = 1
            elif 'month' in age[1]:
                df.at[cell,column] = round(int(age[0]) * (365/12) / 7)
            elif 'year' in age[1]:
                df.at[cell,column] = int(age[0]) * 52
            elif 'week' in age[1]:
                df.at[cell,column] = int(age[0])
            elif 'day' in age[1]:
                df.at[cell,column] = (int(age[0]) // 7) if int(age[0]) > 7 else 1                 
            else:
                df.at[cell,column] = 0
    df[column] = df[column].astype('int64')
```

**Data Entry has a sense of inaccuracy with respect to resolution of days-weeek-month-year. This hurdle is mentioned in data quality report and addressed in data Quality plan**
 - For data processing, unit of a continous feature must be consistent.
 - Hence in above step, month and year data is converted into weeks.
 - But Data entry in Months and Year does not have enough granularity.
     - eg. For animal A783721, mentioned features are DateTime_Intake = 11/3/2018  12:32:00 PM; Date of Birth = 9/5/2018. This makes age of animal concerned equal to 9 weeks at time of Intake. <br> But since data entry at time of intake is in terms of months, and the animal is 2 days short to age of 2 months, Data entry is 1 month. This error is propogated in above step which resolves months into weeks. 
 - Very same logic can be proved for animals with age more than an year and age in weeks. 
     - eg. For animal A788556, which is of age 1 year, 2 months, 1 week at time of intake; data entry is simply 1 year.
     - eg. For animal A791657, which is of age 19 days, at time of intake; data entry is simply 2 weeks.     


```python
#DATETIME DATA

# Select columns containing datetime data
continous_date_columns = df[['DateTime_Intake', 'MonthYear_Intake', 'DateTime_Outcome', 'MonthYear_Outcome']].columns

# Assign object type datetime to columns enlisted in continous_date_columns
for column in continous_date_columns:
    df[column] = pd.to_datetime(df[column], format="%m/%d/%Y %I:%M:%S %p")

df['Date of Birth'] = pd.to_datetime(df['Date of Birth'], format="%m/%d/%Y")
```

# 2. Data Quality Plan

## 2.1 Data findings  

### 2.1.1 Summary of list of issues found in Data Quality Report  

- Datetime of intake is greater for few samples than datetime of outcome due to interchanged data entry
- Age upon intake is greater for few samples than Age upon outcome due to interchanged data entry
- Age upon intake and Age upon outcome are imprecise given reasons explained in [1.1.3.4](####1.1.3.4-Convert-to-decided-data-type)
- Color upon intake and breed upon intake features have too many catagories for given sample size.
- Name_Intake feature has few entries with leading '*', few bad entries starting with numerics

### 2.1.2 Proposed solutions to rectify identified problems   

- Interchange respected values for Date of intake and Date of Outcome
- Resolve errors for Age upon intake and Age upon Outcome due to data entry unit discripancy errors.
- Keep only major sections for category name for Breed_Intake and Color_Intake features
- Strip prefixed '*', and replace invalid entries and null values with string "Unknown" for feature "Name_Intake"

## 2.2 Apply solutions to address data quality issues

### 2.2.1 DateTime Intake > DateTime Outcome

```python
# List all relevent columns of df to observe bad data entry regarding DateTime Intake and DateTime Outcome
dTest = df[["Animal ID","Date of Birth", "DateTime_Intake","Age upon Intake","DateTime_Outcome","Age upon Outcome"]][df["DateTime_Intake"] > df["DateTime_Outcome"]]
dTest
```

**Though lay man observation; it is evident that data entry of following columns has been interchanged for few samples:**
 - DateTime_Intake <-> DateTime_Outcome
 - Age upon Intake <-> Age upon Outcome
 
To substatiate this claim, actual time difference in weeks is calculated for pairs [DateTime_Intake,Date of Birth] and [DateTime_Outcome,Date of Birth]

```python
# Compare data entry for Ages upon intake and outcome with atcual calculations of "Age upon Intake" and "Age upon Outcome" 

dTest_1 = dTest[["Age upon Intake","Age upon Outcome"]].join(\
        [dTest["DateTime_Intake"].sub(df["Date of Birth"], axis=0) // np.timedelta64(1, 'W'),\
         dTest["DateTime_Outcome"].sub(df["Date of Birth"], axis=0) // np.timedelta64(1, 'W')])

dTest_1.columns = ["Age upon Intake","Age upon Outcome","Calculated_AgeIntake","Calculated_AgeOutcome"]
dTest_1
```

It is evident that for all cases, 
 - Age upon Intake <= Calculated_AgeIntake
 - Age upon Outcome <= Calculated_AgeOutcome 
 
**Follwing column entries are interchanged for few columns:**
 - DateTime_Intake <-> DateTime_Outcome
 - Age upon Intake <-> Age upon Outcome
 <br><br>
 
**Interchanging sample entries in error**

```python
for row in dTest.index:
    df.at[row, 'Age upon Intake'],df.at[row, 'Age upon Outcome'] = df.at[row, 'Age upon Outcome'], df.at[row, 'Age upon Intake']
    df.at[row, 'DateTime_Intake'],df.at[row, 'DateTime_Outcome'] = df.at[row, 'DateTime_Outcome'], df.at[row, 'DateTime_Intake']
```

**Running test_3 to varify the data cleaning**

```python
test_3_1 = df[["DateTime_Intake","DateTime_Outcome"]][df["DateTime_Intake"] > df["DateTime_Outcome"]]
print("Number of rows failing the test: ", test_3_1.shape[0])
test_3_1
```

### 2.2.5 Data wrangling for Feature Color

Animal color has very large Unique Entry: Total sample size ratio  for a catagorical datatype. And this granularity does not seem a decisive factor in target variable, because -
 - Most animals have a prominant colour and 2-3 patch colors.
 - hence, a cat being black smoke or black tabby does not add any excess information for this sample size. It can simply be refered as black.
<br>

**Hence animal colors are abstracted to field value preceding seperator '/'** <br>
eg. "Orange Tabby/White" is treated as --->  "Orange Tabby"

```python
df['Color_Intake'] = df['Color_Intake'].str.split('/').str[0]
df['Color_Intake'] = df['Color_Intake'].astype('category')
```

### 2.2.6 Data wrangling for Feature Breed

Animal breed has very large Unique Entry: Total sample size ratio for a catagorical datatype. And this granularity does not seem a decisive factor in target variable, because -
 - Most animals have a base breed followed by its variety.
 - hence, a cat being Labrador Retriever/Anatol Shepherd or Labrador Retriever/Dachshund does not add any excess information for this sample size. It can simply be refered as Labrador Retriever.
<br>

**Hence animal breed are abstracted to field value preceding seperator '/'** <br>
eg. "Pit Bull/Australian Cattle Dog" is treated as --->  "Pit Bull"

```python
df['Breed_Intake'] = df['Breed_Intake'].str.split('/').str[0]
df['Breed_Intake'] = df['Breed_Intake'].astype('category')
```

### 2.2.7 Data wrangling for feature Name_Intake 

```python
# Strip leading '*' in the name
df['Name_Intake'] = df['Name_Intake'].str.lstrip('*')

# Replace all NaN with name "Unknown"
df['Name_Intake'] = df['Name_Intake'].fillna('Unknown')

#Replace all non name emtries starting with a number with name "Unknown" 
# non name entries exist on index 712('1'),713('250 Grams'),400('156 Grams'),819('355 Grams')
df['Name_Intake'][712] = "Unknown"
df['Name_Intake'][713] = "Unknown"
df['Name_Intake'][400] = "Unknown"
df['Name_Intake'][819] = "Unknown"

df['Name_Intake'] = df['Name_Intake'].astype('category')
```

# 4. Feature Engineering

## 4.1 new feature set 1

Castration can affect male and feamle differnetly. It is also important factor for them getting adopted.
 - Hence dissolving feature Sex upon Intake into Sex_Intake and Sex_State_Intake can lead to importnt discoveries. 
 - Category "Unknown" is preserved.

```python
# Entries with category "Unknown" is padded with string " Unknown" for operational purpose
df['Sex upon Intake'] = df['Sex upon Intake'].astype(str)

for row in df.index:
    gender = df.at[row, 'Sex upon Intake']
    
    df.at[row, 'Sex upon Intake']= "Unknown Unknown" if (gender == "Unknown") else gender
```

```python
gender = df['Sex upon Intake'].str.split(' ')

Sex_Intake = np.array(gender.str[0])
Sex_State_Intake = np.array(gender.str[1])

df['Sex_Intake'] = Sex_Intake
df['Sex_State_Intake'] = Sex_State_Intake

df['Sex_Intake'] = df['Sex_Intake'].astype('category')
df['Sex_State_Intake'] = df['Sex_State_Intake'].astype('category')

# df.drop(['Sex upon Intake'], axis=1)
del df['Sex upon Intake']
```

Castration can affect male and feamle differnetly. It is also important factor for them getting adopted.
 - Hence dissolving feature Sex upon Outcome into Sex_Outcome and Sex_State_Outcome can lead to importnt discoveries.
 - Category "Unknown" is preserved.

```python
# Entries with category "Unknown" is padded with string " Unknown" for operational purpose
df['Sex upon Outcome'] = df['Sex upon Outcome'].astype(str)

for row in df.index:
    gender = df.at[row, 'Sex upon Outcome']
    
    df.at[row, 'Sex upon Outcome']= "Unknown Unknown" if (gender == "Unknown") else gender
```

```python
gender = df['Sex upon Outcome'].str.split(' ')

Sex_Outcome = np.array(gender.str[0])
Sex_State_Outcome = np.array(gender.str[1])

df['Sex_Outcome'] = Sex_Outcome
df['Sex_State_Outcome'] = Sex_State_Outcome

df['Sex_Outcome'] = df['Sex_Outcome'].astype('category')
df['Sex_State_Outcome'] = df['Sex_State_Outcome'].astype('category')

# df.drop(['Sex upon Outcome'], axis=1)
del df['Sex upon Outcome']
```

## 4.3 new feature set 3

Locality plays vital role in finding owners of an animal. It also dictates animals health at time of intake. Hence classifying the "Found Location" into constituent features "found_Address_Line1", "found_Address_City" and "found_Address_State" can be fruitful.<br>

Standard data entry is in format "<address line 1> in <city> [<state>]". 
 - Yet some outliers have only city name follwed by state. Such entries are prepended by string "Unknown address in". 
 - Special values "Outside Jurisdiction" are also appended by same string.

```python
# Prepend fields with no address line with string "Unknown address in"


df['Found Location'] = df['Found Location'].astype(str)
addString = "Unknown address in "
for row in df.index:
    location = str(df.at[row, 'Found Location'])
    if not (location.find(" in ") > -1):
        df.at[row, 'Found Location'] = str(addString+location)
    else:
        df.at[row, 'Found Location'] = location
```

```python
location = df['Found Location'].str.split(' in ')

found_Address_Line1 = np.array(location.str[0])

location = location.str[1]
cityState = []
found_Address_City = []
found_Address_State = []
for loc in location:
    cityState = loc.split(" ")
    found_Address_City.append(cityState[0])
    found_Address_State.append(cityState[1].strip('()'))

df['found_Address_Line1'] = found_Address_Line1
df['found_Address_City'] = found_Address_City
df['found_Address_State'] = found_Address_State

df['found_Address_Line1'] = df['found_Address_Line1'].astype('category')
df['found_Address_City'] = df['found_Address_City'].astype('category')
df['found_Address_State'] = df['found_Address_State'].astype('category')

# df.drop(['Found Location'], axis=1)
del df['Found Location']
```

## 4.4 new feature 4

**recidence_Time_Ratio** <br>
- Animal's outside condition in wild/ stray/ owned stage hugely impacts its life at shelter. Captive animals easily repsond to treatements, adjust with peers etc. Whereas, it is difficult for wild animals and their survival rate can be lesser than captive animals. This is evident from outcome seen in [3.1](##3.1-Catagorical-vs-Catagorical) <br>
- Thus ratio of time spent by animal in outside world to the time spent by it in shelter can give us a measure of its survial tendancy.<br>
- This ratio is calculated in days for precision.<br>

$ Target\ Variable\ \alpha \ \ \ \frac{\large Time\ in\ Outer\ World}{\large Time\ in\ Shelter} $<br>

where,<br>
$ Time\ in\ outer\ world \alpha \ \ \ (Date\ of\ Birth\ - Date\ of\ Intake) $<br>
$ Time\ in\ shelter \alpha \ \ \ (Date\ of\ Outcome - Date\ of\ Intake)$<br>

```python
df['recidence_Time_Ratio'] = round(((df["DateTime_Intake"].sub(df["Date of Birth"], axis=0) // np.timedelta64(1, 'D'))+1)\
                                / (df["DateTime_Outcome"].sub(df["DateTime_Intake"], axis=0) // np.timedelta64(1, 'D') + 1))

df['recidence_Time_Ratio'] = df['recidence_Time_Ratio'].astype('int64')
```

```python
df[df['binary_outcome'] == 1][['recidence_Time_Ratio']].plot(kind='hist',figsize=(7,7), bins=10, title="recidence_Time_Ratio marker")
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
```

**This is a more prominant measure of outcome as compared to that of [3.2](##3.2-Continuous-vs-Catagorical)**
